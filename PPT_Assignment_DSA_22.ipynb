Given a Binary Tree (Bt), convert it to a Doubly Linked List(DLL). The left and right pointers in nodes are to be used as previous and next pointers respectively in converted DLL. The order of nodes in DLL must be the same as in Inorder for the given Binary Tree. The first node of Inorder traversal (leftmost node in BT) must be the head node of the DLL.



</aside>

```python
class Node: 
    def __init__(self, data): 
        self.data = data  
        self.left = None
        self.right = None
  
def BinaryToDLL(root): 
    if root is None: 
        return root 
  
    if root.left: 
        left = BinaryToDLL(root.left) 
  
        while left.right: 
            left = left.right 
  
        left.right = root 
        root.left = left 
  
    if root.right: 
        right = BinaryToDLL(root.right) 
  
        while right.left: 
            right = right.left 
  
        right.left = root 
        root.right = right 
  
    return root 

def BTtoDLL(root): 
    if root is None: 
        return root 
  
    root = BinaryToDLL(root) 
  
    while root.left: 
        root = root.left 
  
    return root
# Construct the binary tree
root = Node(10) 
root.left = Node(12) 
root.right = Node(15) 
root.left.left = Node(25) 
root.left.right = Node(30) 
root.right.left = Node(36) 

# Convert to DLL 
root = BTtoDLL(root) 

# Print the converted list
node = root
while node:
    print(node.data, end=" ")
    node = node.right
```

A Given a binary tree, the task is to flip the binary tree towards the right direction that is clockwise. See the below examples to see the transformation.

In the flip operation, the leftmost node becomes the root of the flipped tree and its parent becomes its right child and the right sibling becomes its left child and the same should be done for all left most nodes recursively.


</aside>

```python
# Python3 program to flip
# a binary tree
from collections import deque

# A binary tree node structure
class Node:

	def __init__(self, key):
	
		self.data = key
		self.left = None
		self.right = None

# method to flip the
# binary tree
def flipBinaryTree(root):

	# Initialization of
	# pointers
	curr = root
	next = None
	temp = None
	prev = None

	# Iterate through all
	# left nodes
	while(curr):
		next = curr.left

		# Swapping nodes now, need temp
		# to keep the previous right child

		# Making prev's right as curr's
		# left child
		curr.left = temp

		# Storing curr's right child
		temp = curr.right

		# Making prev as curr's right
		# child
		curr.right = prev

		prev = curr
		curr = next
	return prev

# Iterative method to do level
# order traversal line by line
def printLevelOrder(root):

	# Base Case
	if (root == None):
		return

	# Create an empty queue for
	# level order traversal
	q = deque()

	# Enqueue Root and initialize
	# height
	q.append(root)

	while (1):
		# nodeCount (queue size) indicates
		# number of nodes at current level.
		nodeCount = len(q)
		if (nodeCount == 0):
			break

		# Dequeue all nodes of current
		# level and Enqueue all nodes
		# of next level
		while (nodeCount > 0):
			node = q.popleft()
			print(node.data, end = " ")

			if (node.left != None):
				q.append(node.left)

			if (node.right != None):
				q.append(node.right)
			nodeCount -= 1

		print()

# Driver code
if __name__ == '__main__':

	root = Node(1)
	root.left = Node(2)
	root.right = Node(3)
	root.right.left = Node(4)
	root.right.right = Node(5)

	print("Level order traversal of given tree")
	printLevelOrder(root)

	root = flipBinaryTree(root)

	print("\nLevel order traversal of the flipped"
		" tree")
	printLevelOrder(root)

# This code is contributed by Mohit Kumar 29
```

ðŸ’¡ Question-3:

Given a binary tree, print all its root-to-leaf paths without using recursion. For example, consider the following Binary Tree.

Input:

        6
     /    \
    3      5
  /   \     \
 2     5     4
     /   \
    7     4

Output:

There are 4 leaves, hence 4 root to leaf paths -
  6->3->2
  6->3->5->7
  6->3->5->4
  6->5>4

</aside>

```python
class Node: 
    def __init__(self, data): 
        self.data = data 
        self.left = None
        self.right = None

def printRootToLeafPaths(root): 
    if not root: 
        return

    # Stack to store the nodes
    nodes = []
    nodes.append(root)

    # Stack to store the paths leading to the nodes
    paths = []
    paths.append(str(root.data))

    while len(nodes) > 0:
        node = nodes.pop()
        path = paths.pop()

        if node.left is None and node.right is None:
            # We have reached a leaf node, print the path
            print(path)

        if node.right is not None:
            nodes.append(node.right)
            paths.append(path + '->' + str(node.right.data))

        if node.left is not None:
            nodes.append(node.left)
            paths.append(path + '->' + str(node.left.data))

# Create binary tree
root = Node(6)
root.left = Node(3)
root.right = Node(5)
root.left.left = Node(2)
root.left.right = Node(5)
root.left.right.left = Node(7)
root.left.right.right = Node(4)
root.right.right = Node(4)

# Print root-to-leaf paths
printRootToLeafPaths(root)
```

GivenÂ Preorder, Inorder and PostorderÂ traversals of some tree. Write a program to check if they all are of the same tree.

**Examples:**

Input : 

        Inorder -> 4 2 5 1 3
        Preorder -> 1 2 4 5 3
        Postorder -> 4 5 2 3 1
Output : 

Yes
Explanation : 

All of the above three traversals are of
the same tree 

                           1
                         /   \
                        2     3
                      /   \
                     4     5

Input : 

        Inorder -> 4 2 5 1 3
        Preorder -> 1 5 4 2 3
        Postorder -> 4 1 2 3 5
Output : 

No

</aside>

```python
def checktree(pre_order, in_order, post_order, len): 

    if len == 0: 
        return True
  
    if len == 1: 
        return (pre_order[0] == in_order[0]) and (in_order[0] == post_order[0])

    idx = -1

    for i in range(len): 
        if (in_order[i] == pre_order[0]): 
            idx = i 
            break

    if(idx == -1): 
        return False

    if pre_order[0] != post_order[len-1]:
        return False

    return (checktree(pre_order[1:idx+1], in_order[:idx], post_order[:idx], idx) and
            checktree(pre_order[idx+1:], in_order[idx+1:], post_order[idx:len-1], len-idx-1))


def check(pre_order, in_order, post_order, len): 

    if (checktree(pre_order, in_order, post_order, len) == True): 
        return "Yes"
    return "No"

in_order = [4, 2, 5, 1, 3]
pre_order = [1, 2, 4, 5, 3]
post_order = [4, 5, 2, 3, 1]

print(check(pre_order, in_order, post_order, len(in_order)))  # Should print 'Yes'

in_order = [4, 2, 5, 1, 3]
pre_order = [1, 5, 4, 2, 3]
post_order = [4, 1, 2, 3, 5]

print(check(pre_order, in_order, post_order, len(in_order)))  # Should print 'No'
```
