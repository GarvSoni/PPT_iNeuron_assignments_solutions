Given an array, for each element find the value of the nearest element to the right which is having a frequency greater than that of the current element. If there does not exist an answer for a position, then make the value â€˜-1â€™.

**Examples:**
Input: a[] = [1, 1, 2, 3, 4, 2, 1]
Output : [-1, -1, 1, 2, 2, 1, -1]

Explanation:
Given array a[] = [1, 1, 2, 3, 4, 2, 1]
Frequency of each element is: 3, 3, 2, 1, 1, 2, 3

Lets calls Next Greater Frequency element as NGF
1. For element a[0] = 1 which has a frequency = 3,
   As it has frequency of 3 and no other next element
   has frequency more than 3 so  '-1'
2. For element a[1] = 1 it will be -1 same logic
   like a[0]
3. For element a[2] = 2 which has frequency = 2,
   NGF element is 1 at position = 6  with frequency
   of 3 > 2
4. For element a[3] = 3 which has frequency = 1,
   NGF element is 2 at position = 5 with frequency
   of 2 > 1
5. For element a[4] = 4 which has frequency = 1,
   NGF element is 2 at position = 5 with frequency
   of 2 > 1
6. For element a[5] = 2 which has frequency = 2,
   NGF element is 1 at position = 6 with frequency
   of 3 > 2
7. For element a[6] = 1 there is no element to its
   right, hence -1
</aside>

```python
def findNearestGreaterFrequency(arr):
    n = len(arr)
    frequency = {}
    nextGreaterFreq = {}
    result = []

    # Calculate frequency of each element
    for num in arr:
        frequency[num] = frequency.get(num, 0) + 1

    stack = []
    for i in range(n-1, -1, -1):
        while stack and frequency[arr[i]] >= frequency[stack[-1]]:
            stack.pop()

        if stack:
            nextGreaterFreq[arr[i]] = stack[-1]
        else:
            nextGreaterFreq[arr[i]] = -1

        stack.append(arr[i])

    # Create the result array using nextGreaterFreq dictionary
    for num in arr:
        result.append(nextGreaterFreq[num])

    return result



arr = [1, 1, 1, 2, 2, 2, 2, 11, 3, 3]
print("\nInput:", arr)
print("Output:", findNearestGreaterFrequency(arr))
```

ðŸ’¡ **Question 2**

Given a stack of integers, sort it in ascending order using another temporary stack.

**Examples:**
Input : [34, 3, 31, 98, 92, 23]
Output : [3, 23, 31, 34, 92, 98]

Input : [3, 5, 1, 4, 2, 8]
Output : [1, 2, 3, 4, 5, 8]
</aside>

```python
def sortStack(stack):
    if not stack:
        return stack

    tempStack = []
    while stack:
        temp = stack.pop()
        while tempStack and tempStack[-1] > temp:
            stack.append(tempStack.pop())
        tempStack.append(temp)

    return tempStack

# Test the implementation
stack = [34, 3, 31, 98, 92, 23]
print("Input Stack:", stack)
sortedStack = sortStack(stack)
print("Sorted Stack:", sortedStack)

stack = [3, 5, 1, 4, 2, 8]
print("\nInput Stack:", stack)
sortedStack = sortStack(stack)
print("Sorted Stack:", sortedStack)
```

Input Â :Â Stack[] = [1, 2, 3, 4, 5]

Input Â :Â Stack[] = [1, 2, 3, 4, 5, 6]

Output :Â Stack[] = [1, 2, 4, 5, 6]

</aside>

```python
def deleteMiddle(stack, n, current):
    if current == n // 2 + 1:
        stack.pop()
        return

    temp = stack.pop()
    deleteMiddle(stack, n, current+1)
    stack.append(temp)

def deleteMiddleElement(stack):
    n = len(stack)
    if n == 0:
        return stack

    deleteMiddle(stack, n, 1)
    return stack

# Test the implementation
stack = [1, 2, 3, 4, 5]
print("Original Stack:", stack)
stack = deleteMiddleElement(stack)
print("Stack after deleting middle element:", stack)

stack = [1, 2, 3, 4, 5, 6]
print("\nOriginal Stack:", stack)
stack = deleteMiddleElement(stack)
print("Stack after deleting middle element:", stack)
```

1. Push and pop elements from the stack
2. Pop (Or Dequeue) from the given Queue.
3. Push (Or Enqueue) in the another Queue.

**Examples :**

Input :Â Queue[] = { 5, 1, 2, 3, 4 }Â 

Output :Â YesÂ 

Pop the first element of the given Queue 

i.e 5.Â Push 5 into the stack.Â 

Now, pop all the elements of the given Queue and push them toÂ second Queue.Â 

Now, pop element 5 in the stack and push it to the second Queue.Â Â Â

Output :Â NoÂ 

Push 5 to stack.Â 

Pop 1, 2 from given Queue and push it to another Queue.Â 

Pop 6 from given Queue and push to stack.Â 

Pop 3, 4 from given Queue and push to second Queue.Â 

Now, from using any of above operation, we cannot push 5Â into the second Queue because it is below the 6 in the stack.

```python
def checkArrangement(queue):
    stack = []
    n = len(queue)
    expected = 1

    for i in range(n):
        if queue[i] == expected:
            expected += 1
        elif stack and stack[-1] == expected:
            expected += 1
            stack.pop()
        else:
            stack.append(queue[i])

    while stack:
        if stack[-1] == expected:
            expected += 1
            stack.pop()
        else:
            return False

    return True

# Test the implementation
queue = [5, 1, 2, 3, 4]
print("Queue:", queue)
print("Can be arranged in increasing order:", checkArrangement(queue))

queue = [5, 1, 2, 6, 3, 4]
print("\nQueue:", queue)
print("Can be arranged in increasing order:", checkArrangement(queue))
```

ðŸ’¡ **Question 5**

Given a number , write a program to reverse this number using stack.

**Examples:**
Input : 365
Output : 563

Input : 6899
Output : 9986
</aside>

```python
def reverseNumber(num):
    stack = []
    while num > 0:
        stack.append(num % 10)
        num //= 10

    reversed_num = 0
    power = 0
    while stack:
        reversed_num += stack.pop() * (10 ** power)
        power += 1

    return reversed_num

# Test the implementation
num = 365
print("Input Number:", num)
print("Reversed Number:", reverseNumber(num))

num = 6899
print("\nInput Number:", num)
print("Reversed Number:", reverseNumber(num))
```

ðŸ’¡ **Question 6**

Given an integer k and aÂ **[queue](https://www.geeksforgeeks.org/queue-data-structure/)**Â of integers, The task is to reverse the order of the firstÂ **k**Â elements of the queue, leaving the other elements in the same relative order.

Only following standard operations are allowed on queue.

- **enqueue(x) :**Â Add an item x to rear of queue
- **dequeue() :**Â Remove an item from front of queue
- **size() :**Â Returns number of elements in queue.
- **front() :**Â Finds front item.
</aside>

```python
from collections import deque

def reverseKElements(queue, k):
    stack = []
    for _ in range(k):
        stack.append(queue.popleft())

    while stack:
        queue.append(stack.pop())

    for _ in range(len(queue) - k):
        queue.append(queue.popleft())

    return queue

# Test the implementation
queue = deque([1, 2, 3, 4, 5])
k = 3
print("Original Queue:", queue)
reversed_queue = reverseKElements(queue, k)
print("Reversed Queue:", reversed_queue)

queue = deque([10, 20, 30, 40, 50, 60])
k = 5
print("\nOriginal Queue:", queue)
reversed_queue = reverseKElements(queue, k)
print("Reversed Queue:", reversed_queue)
```

Given a sequence of n strings, the task is to check if any two similar words come together and then destroy each other then print the number of words left in the sequence after this pairwise destruction.

**Examples:**

InputÂ : ab aa aa bcd ab

OutputÂ : 3

*As aa, aa destroys each other so,*

*ab bcd ab is the new sequence.*

Input : Â tom jerry jerry tom

Output :Â 0

*As first both jerry will destroy each other.*

*Then sequence will be tom, tom they will also destroy*

*each other. So, the final sequence doesnâ€™t contain any*

*word.*

</aside>

```python
def pairwiseDestruction(words):
    stack = []
    for word in words:
        if stack and stack[-1] == word:
            stack.pop()
        else:
            stack.append(word)
    return len(stack)

# Test the implementation
words = ['ab', 'aa', 'aa', 'bcd', 'ab']
print("Number of Words Left:", pairwiseDestruction(words))

words = ['tom', 'jerry', 'jerry', 'tom']
print("\nNumber of Words Left:", pairwiseDestruction(words))
```

Given an array of integers, the task is to find the maximum absolute difference between the nearest left and the right smaller element of every element in the array.

**Note:**Â If there is no smaller element on right side or left side of any element then we take zero as the smaller element. For example for the leftmost element, the nearest smaller element on the left side is considered as 0. Similarly, for rightmost elements, the smaller element on the right side is considered as 0.

**Examples:**
Input : arr[] = {2, 1, 8}
Output : 1
Left smaller  LS[] {0, 0, 1}
Right smaller RS[] {1, 0, 0}
Maximum Diff of abs(LS[i] - RS[i]) = 1

Input  : arr[] = {2, 4, 8, 7, 7, 9, 3}
Output : 4
Left smaller   LS[] = {0, 2, 4, 4, 4, 7, 2}
Right smaller  RS[] = {0, 3, 7, 3, 3, 3, 0}
Maximum Diff of abs(LS[i] - RS[i]) = 7 - 3 = 4

Input : arr[] = {5, 1, 9, 2, 5, 1, 7}
Output : 1
</aside>

```python
def maxAbsoluteDifference(arr):
    n = len(arr)
    left_smaller = [0] * n
    right_smaller = [0] * n

    # Find nearest smaller element on the left side
    stack = []
    for i in range(n):
        while stack and stack[-1] >= arr[i]:
            stack.pop()
        if stack:
            left_smaller[i] = stack[-1]
        stack.append(arr[i])

    # Find nearest smaller element on the right side
    stack = []
    for i in range(n-1, -1, -1):
        while stack and stack[-1] >= arr[i]:
            stack.pop()
        if stack:
            right_smaller[i] = stack[-1]
        stack.append(arr[i])

    max_diff = 0
    for i in range(n):
        diff = abs(left_smaller[i] - right_smaller[i])
        max_diff = max(max_diff, diff)

    return max_diff

# Test the implementation
arr = [2, 1, 8]
print("Input Array:", arr)
print("Maximum Absolute Difference:", maxAbsoluteDifference(arr))
```
