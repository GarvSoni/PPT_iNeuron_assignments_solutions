Given preorder of a binary tree, calculate itsÂ **[depth(or height)](https://www.geeksforgeeks.org/write-a-c-program-to-find-the-maximum-depth-or-height-of-a-tree/)**Â [starting from depth 0]. The preorder is given as a string with two possible characters.

1. â€˜lâ€™ denotes the leaf
2. â€˜nâ€™ denotes internal node

The given tree can be seen as a full binary tree where every node has 0 or two children. The two children of a node can â€˜nâ€™ or â€˜lâ€™ or mix of both.

**Examples :**

Input  : nlnll
Output : 2

</aside>

```python
def calculateDepth(pre_order, n, index):

    # Tree is empty
    if (index[0] >= n):
        return 0

    if (pre_order[index[0]] == 'l'):  # If the current character is 'l', then it is the leaf node of a full binary tree
        return 1

    index[0] += 1  # Move pointer to next character
    left = calculateDepth(pre_order, n, index)  # Calculate depth of left subtree

    right = calculateDepth(pre_order, n, index)  # Calculate depth of right subtree

    # Return depth as maximum of depths of
    # left and right subtrees plus 1
    return max(left, right) + 1

pre_order = "nlnll"
n = len(pre_order)
index = [0]  # Using list so that we can pass it by reference
print(calculateDepth(pre_order, n, index))  # Output: 2
```

1. â€˜lâ€™ denotes the leaf
2. â€˜nâ€™ denotes internal node

The given tree can be seen as a full binary tree where every node has 0 or two children. The two children of a node can â€˜nâ€™ or â€˜lâ€™ or mix of both.

**Examples :**

Input  : nlnll
Output : 2
Explanation :

Input  : nlnnlll
Output : 3

```python
from collections import deque

class Node:
    def __init__(self, key):
        self.data = key
        self.left = None
        self.right = None

def leftView(root):
    if not root:
        return

    queue = deque()
    queue.append(root)
    while queue:
        n = len(queue)

        for i in range(n):
            temp = queue.popleft()

            if i == 0:
                print(temp.data, end=" ")

            if temp.left:
                queue.append(temp.left)

            if temp.right:
                queue.append(temp.right)

root = Node(4)
root.left = Node(5)
root.right = Node(2)
root.right.left = Node(3)
root.right.right = Node(1)
root.right.left.left = Node(6)
root.right.left.right = Node(7)
leftView(root)  # Output: 4 5 3 6
```

```python
def rightView(root):
    if not root:
        return

    queue = deque()
    queue.append(root)
    while queue:
        n = len(queue)

        for i in range(n):
            temp = queue.popleft()

            if i == n-1:
                print(temp.data, end=" ")

            if temp.left:
                queue.append(temp.left)

            if temp.right:
                queue.append(temp.right)

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)
root.right.right.right = Node(8)
rightView(root)  # Output: 1 3 7 8
```

Given a Binary Tree, The task is to print theÂ **bottom view**Â from left to right. A nodeÂ **x**Â is there in output if x is the bottommost node at its horizontal distance. The horizontal distance of the left child of a node x is equal to a horizontal distance of x minus 1, and that of a right child is the horizontal distance of x plus 1.

**Examples:**

**Input:**

             20

           / Â  Â  \

        8 Â  Â  Â  Â  22

    / Â  Â  Â \ Â  Â  Â  Â  \

5 Â  Â  Â  Â  3 Â  Â  Â  Â 25

        / Â  Â \

   10 Â  Â  Â  14

**Output:**Â 5, 10, 3, 14, 25.

**Input:**

             20

           / Â  Â  \

        8 Â  Â  Â  Â  22

    / Â  Â  Â \ Â  Â  Â / Â  \

 5 Â  Â  Â  Â  3 Â   4 Â  Â  25

         / Â  Â \

     10 Â  Â  Â  14

**Output:**

5 10 4 14 25.

**Explanation:**

If there are multiple bottom-most nodes for a horizontal distance from the root, then print the later one in the level traversal.

**3 and 4** are both the bottom-most nodes at a horizontal distance of 0, we need to print 4.

</aside>

```python
from collections import deque

def bottomView(root):
    if not root:
        return

    hd_node_dict = dict()
    queue = deque()

    hd_node_dict[0] = root.data
    queue.append((root, 0))

    while queue:
        node, hd = queue.popleft()

        hd_node_dict[hd] = node.data

        if node.left:
            queue.append((node.left, hd - 1))

        if node.right:
            queue.append((node.right, hd + 1))

    for i in sorted(hd_node_dict.keys()):
        print(hd_node_dict[i], end=" ")

root = Node(20)
root.left = Node(8)
root.right = Node(22)
root.left.left = Node(5)
root.left.right = Node(3)
root.right.left = Node(4)
root.right.right = Node(25)
root.left.right.left = Node(10)
root.left.right.right = Node(14)
bottomView(root)  # Output: 5 10 4 14 25
```
